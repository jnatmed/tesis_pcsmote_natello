```python
import pandas as pd
import os

# Definir las rutas
ruta_original = 'predictive_maintenance.csv'
ruta_salida = 'predictive_maintenance_bin.csv'

# 1. Cargar el dataset
df = pd.read_csv(ruta_original)

# 2. Convertir a clase binaria
# Asumiendo que la columna se llama 'Failure Type' o similar (según se ve en tu captura)
# Si la columna final es la que quieres modificar:
columna_target = df.columns[-1] 

df[columna_target] = df[columna_target].apply(lambda x: 'No Failure' if x == 'No Failure' else 'Failure')

# 3. Verificar el cambio
print("Distribución de clases:")
print(df[columna_target].value_counts())

# 4. Guardar el nuevo dataset
df.to_csv(ruta_salida, index=False)
print(f"\nArchivo guardado exitosamente en: {ruta_salida}")
```
